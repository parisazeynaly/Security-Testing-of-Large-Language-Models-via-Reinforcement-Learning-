```python
import pandas as pd
import textwrap
from pathlib import Path

# Adjust path if needed
DATA_PATH = Path("/content/coat_input_advbench.csv")

df = pd.read_csv(DATA_PATH)
df.head(), df["y"].value_counts()
```

```python
N_PER_GROUP = 25

def build_data_block(df, n_per_group=25):
    lines = ["# DATA"]
    for level in ["Low", "Medium", "High"]:
        sub = df[df["y"] == level].sample(
            n=min(n_per_group, len(df[df["y"] == level])),
            random_state=42
        )
        lines.append(f"\n## Reward level: {level}\n")
        for _, row in sub.iterrows():
            txt = str(row["text"]).replace("\n", " ")
            txt = " ".join(txt.split())  # normalize spaces
            lines.append(f"- {txt}")
            lines.append("")  # blank line for readability
    return "\n".join(lines)

data_block = build_data_block(df, N_PER_GROUP)
print(data_block[:2000])  # preview
```
